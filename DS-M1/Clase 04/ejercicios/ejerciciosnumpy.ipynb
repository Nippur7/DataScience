Reemplazar ambos valores de las diagonales en 0

```python
import pandas as pd
import numpy as np

df= pd.DataFrame(np.random.randint(1,100,100).reshape(10, -1))
df
```

```python
for i in range(df.shape[0]):
    df.iat[i, i] = 0
    df.iat[df.shape[0]-i-1, i] = 0
```

```python
df
```

How to get the particular group of a groupby dataframe by key?

```python
df = pd.DataFrame({'fruta':['manzana', 'banana', 'naranja'] * 3,
                    'rating': np.random.rand(9),
                    'precio': np.random.randint(0,15,9)})
```

```python
df
```

```python
df_agrup = df.groupby(['fruit'])
df_agrup.get_group('apple')
```

```python
salida = df.groupby('fruta', as_index=False)['precio'].mean()
print(salida)
```

How to get the positions where values of two columns match?

```python
df = pd.DataFrame({'fruit1': np.random.choice(['apple', 'orange', 'banana'], 10),
                    'fruit2': np.random.choice(['apple', 'orange', 'banana'], 10)})
df
```

```python
np.where(df.fruit1 == df.fruit2)
```

How to compute the correlation of each row with the suceeding row?

```python
df = pd.DataFrame(np.random.randint(1,100, 80).reshape(8, -1))
df
```

```python
[df.iloc[i].corr(df.iloc[i+1]).round(2) for i in range(df.shape[0])[:-1]]
```

```python
How to get the n’th largest value of a column when grouped by another column?
```

```python
df = pd.DataFrame({'fruit': ['apple', 'banana', 'orange'] * 3,
                   'taste': np.random.rand(9),
                   'price': np.random.randint(0, 15, 9)})
df
```

```python
grupo = df['price'].groupby(df.fruit)
grupo.get_group('apple').sort_values().iloc[-2]
```
