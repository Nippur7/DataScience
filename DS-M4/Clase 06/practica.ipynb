```python
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb://localhost:27017/')
filter={
    'properties.BARRIO': {
        'pattern': 'Barracas', 
        'options': ''
    }
}
sort=list({
    'properties.OBJECTID': -1
}.items())

result = client['tp_mongo']['CMB'].find(filter=filter,sort=sort)
for res in result:
    pprint(res)
```

```python
from pymongo import MongoClient

# Requires the PyMongo package.
# https://api.mongodb.com/python/current

client = MongoClient('mongodb://localhost:27017/')
filter={}
project={
    'properties.NOMBRE': 1, 
    'properties.Dom_mapa': 1, 
    'geometry.coordinates': 1
}

result = client['tp_mongo']['CMB'].find(
  filter=filter,
  projection=project
)

#for res in result:
#    pprint(res)
salida = open('centros.csv', 'w', encoding='utf-8')
salida.write('latitude,longitude,name,label\n')
for res in result:
  nombre =res['properties']['NOMBRE']
  direccion = res['properties']['Dom_mapa']
  lon = res['geometry']['coordinates'][0]
  lat = res['geometry']['coordinates'][1]
  #print(nombre,direccion,lon,lat)
  salida.write(f'{lat},{lon},{direccion},{nombre}\n')

salida.close()
```
